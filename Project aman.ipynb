### 18. Data Manipulation.

```python
# removing the null values from the Rating's
def ins(series):
    return series.fillna(series.median())
```

```python
info.Rating=info['Rating'].transform(ins)
```

```python
info.isnull().sum()
# Befor Manipulation The Rating's Has 1474 Null Values, but now it is zero means null values ar filled 
```

## 19. Ploting graph, Category  on X axis & Rating on Y axis

## 21. Plot Bar Graph for the No of Apps in each Catogory.

```python
plt.figure(figsize=(12,5))       # Plot Bar Graph for the No of Apps in each Catogory
plt.title("Category Wise Apps")
plt.ylabel('Number.of Apps')
plt.xlabel('Category')
plt.xticks(rotation=60,fontsize=10)
info['Category'].value_counts().head(10).plot(kind='bar')
plt.show()
```

## 22. Plot Graph for the No of Apps with respect to Android Version.


```python
df_andriod= pd.DataFrame(info.groupby('Android Ver').size(),columns=['Count'])
plt.figure(figsize=(12,8))
plt.title('Andriod versions vs no of apps supported')
df_andriod['Count'].plot(kind='barh')                  # Plot Graph for the No of Apps with respect to Android Version
plt.show()
```

```python
# from the general wise, when i was studying the apps with respect to the andriod version, 
#i could see most apps requires a minimum andriod version of 4.1 and above
#Also andriod version 8 and above was less as the andriod version 8 was latest one during that time
```

```python
plt.title('Variation according to the "Android Version" of the App',fontweight=600)
plt.ylabel('Android Version')
plt.xlabel('Number of Apps')
info['Android Ver'].value_counts().head(10).plot(kind='barh')
plt.show()
```
